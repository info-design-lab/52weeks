```python
import csv
import seaborn
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
import numpy as np
np.random.seed(sum(map(ord, "distributions")))
import pandas as pd
import ast
```

```python
data = []

with open('data.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
```

```python
data[0].keys()
```

```python
# convert datatypes
for d in data:
    for i in d:
        try:
            d[i] = float(d[i])
        except:
            continue
    d['genres'] = ast.literal_eval(d['genres'])
    if 'Horror' in d['genres']:
        d['genres'] = 'Horror'
    elif 'Drama' in d['genres']:
        d['genres'] = 'Drama'
    elif 'Animation' in d['genres']:
        d['genres'] = 'Animation'
    elif 'Action' in d['genres']:
        d['genres'] = 'Action'
    elif 'Documentary' in d['genres']:
        d['genres'] = 'Documentary'
    else:
        d['genres'] = 'Other'
```

```python
data[0]
```

```python
x = [d['popularity'] for d in data]
```

```python
y = [d['revenue'] for d in data]
```

```python
x = np.array(x)
y = np.array(y)

sns.jointplot(x=x, y=y)
```

```python
fields = ['popularity', 'revenue', 'runtime', 'vote_average', 'vote_count', 'budget']
#fields = ['revenue', 'release_date']
```

```python
df = []
for i in data:
    d = {}
    for f in fields:
        d[f] = i[f]
    df.append(d)
```

```python
df = pd.DataFrame(df)
```

```python
#sns.pairplot(df, kind="reg")
sns.pairplot(df)
```
